# **<font color=a66500>What is Python?**
Python is a general-purpose, object-oriented, easy-to-learn and high-level programming language. *Guido van Rossum* designed Python and released it in 1991. He wanted to give it a unique name, so he named it after the famous BBC comedy TV show "Monty Python's Flying Circus".  
  
Python as a language supports simple English-like statements, making it easy for programmers to learn and maintain. This is one important reason for its popularity.

# **<font color=a66500>Statements in Python**
Instructions written in source code and executed by a Python interpreter are called statements. In Python, we can write statements in 3 different ways. Let us learn them in detail.

## **<font color=000000>Simple Statements**
By default, the end of a statement is indicated by pressing the Enter key. Each statement is written on a new line. For example, a=5 is a simple statement where variable 'a' is created with value 5.

```python
a=5
```

## **<font color=000000>Multiline Statements**
In Python, a statement that doesn't fit on oner line can be done in several  ways  

*  **Using Parentheses ():** You cn break a long expression or function call across multiple lines if it is enclosed in parentheses. Python allows this without any additional special characters.

```python
result = (11 + 15 + 148 +
          165 + 26)
```

*  **Using Braces {} (for dictionaries or sets):** Similar to parentheses, curly braces can be used to break long dictionary or set definitions across lines.

```python
my_dict = {
    'name': 'John Doe',
    'age': 30,
    'city': 'New York'
}
```

*  **Using Square Brackets [] (for lists):** You can also break a list definition or list comprehension into multiple lines using square brackets.

```python
my_list = [
    1, 2, 3, 4,
    5, 6, 7, 8
]
```

*  **Using the Continuation Character (\\):** a backslash (\\) at the end of a line indicates that the statement continues on the next line. This method is used for breaking any statements into multiple lines.

```python
result = 23 + 54 + \
45 + 54
```

## **<font color=000000>Multiple Statements**
Multiple Statements means more than one statement in a single line. This can be done by seperating the statements using semicolons (;).

```python
a=5; b=9; c=a+b; print(c)
```
